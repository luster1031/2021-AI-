```python
import numpy as np
import re
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/all_products_v3.csv',encoding='utf-8')
```

```python
data
```

```python
for count in data.index:
  #print(count)
  if type(data._get_value(count,'NM')) == str:
    row_s = data._get_value(count,'NM')
    k_str = ''
    en_str = ''
    add_data = data.iloc[count]
    for c in row_s:
      if ord('가')<= ord(c) <= ord('힣'):
        k_str += c
      elif ord('a') <= ord(c.lower()) <= ord('z'):
        en_str += c
      if ord(' ') == ord(c):
        en_str += ' '
        k_str += ' '
    print(r"문장 : {} \n한국어 : {}".format(c,k_str))
    add_data['NM_EN'] = en_str
    add_data['NM_KO'] = k_str
    data = data.append(add_data,ignore_index=True)
```
