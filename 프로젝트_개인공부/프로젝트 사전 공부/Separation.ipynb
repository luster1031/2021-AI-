```python
import pandas as pd
```

```python
#추가 데이터
add_df = pd.read_csv('./83류.csv',encoding='utf-8')
add_df.columns= ['HSCODE','what','NM','이유','설명']
#특수문자 시작되는거 없애기
dictionary = {'○':'', ' ':'','ㅇ':''}
add_df.replace(dictionary,regex=True, inplace = True)
#한글 삭제
add_df['HSCODE']=add_df['HSCODE'].str.replace(pat=r'[ㄱ-ㅣ가-힣]+', repl= r'', regex=True)
add_df['HSCODE'] = add_df['HSCODE'].str.replace('.','')
#특수문자 시작되는거 없애기
dictionary = {'\n':'', '\t':'','-':''}
add_df.replace(dictionary,regex=True, inplace = True)
add_df['HSCODE'] = add_df['HSCODE'].str.extract(r'(\d\d\d\d\d\d\d\d\d\d)')
add_df = add_df.dropna()
add_df['HSCODE'] = add_df['HSCODE'].astype('int64')
```

```python
import re
reg = re.compile(r'[^ㄱ-ㅣ가-힣]+')

k = 0

for count in add_df.index:
    if type(add_df._get_value(count,'NM')) == str:
        s = add_df._get_value(count,'NM').split(';')
        data = add_df.iloc[count]
        check = False
        for i in range(len(s)-1):
            check = True
            if reg.match(s[i]):#영어
                data['NM_EN'] = s[i]
            else :
                data['NM_KO'] = s[i]
            add_df = add_df.append(data,ignore_index=True)
        if len(s) == 1:
            if reg.match(s[0]):#영어
                data['NM_EN'] = s[0]
            else :
                data['NM_KO'] = s[0]
            add_df = add_df.append(data,ignore_index=True)
```

```python
add_df = add_df.drop(['NM'],axis=1)
```
